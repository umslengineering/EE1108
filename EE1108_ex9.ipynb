````python
# ShowForStrings.py
"""Illustrates the use of for-loops
in problems where it is necessary to
traverse a string."""

def Reverse(s):
    """Returns a string that is obtained
    from s by reversing the order of its characters.

    Precondition: s is a string."""
    t = ''
    for c in s:
        t = c + t
    return t


def nDigits(s):
    """Returns an int whose value is the number
    of digit characters that are in s.
    Precondition: s is a string."""
    n = 0
    for c in s:
        # Increment n if c is a digit
        if c.isdigit():
            n = n + 1
    return n


if __name__ == '__main__':
    print("Using input so enter strings without the apostrophe delimiters.\n")

    x = input('Enter a string to reverse: ')
    y = Reverse(x)
    print('              The reverse: %s' % y)

    x = input('Enter a string with some digits: ')
    m = nDigits(x)
    print('               Number of digits: %1d' % m)
````

````python
# ShowRange
"""Small for-loop examples that illustrate how
range works."""

print('\nrange(5) values:')
for k in range(5):
    print(' ', k)

print('\nrange(0,5) values:')
for k in range(0, 5):
    print(' ', k)

print('\nrange(1,5) values:')
for k in range(1, 5):
    print(' ', k)

print('\nrange(0,5,2) values:')
for k in range(0, 5, 2):
    print(' ', k)

print('\nrange(0,5,-2) values:')
for k in range(0, 5, -2):
    print(' ', k)

print('\nrange(5,0,-2) values:')
for k in range(5, 0, -2):
    print(' ', k)
````

# Distance between Earth to Astroid planet

## Objects
- Earth (E)
- Rogue planet (A)

## Properties
Each has:
- an x-coordinate
- a y-coordinate

---

## Earth Position
```python
xE = 94.5 * math.cos(2 * math.pi * t / 365)
yE = 91.5 * math.sin(2 * math.pi * t / 365)
```

This describes Earth moving along an orbit.

### Why cosine and sine?

Because circular (or elliptical-like) motion can be written as:
```
x = a * cos(angle)
y = b * sin(angle)
```

Where:
- `a` = horizontal radius (semi-major axis)
- `b` = vertical radius (semi-minor axis)

So:
- Earth moves around an ellipse with (unit is million miles):
  - horizontal radius ≈ 94.5
  - vertical radius ≈ 91.5

---

## The Angle
```
2πt / 365
```

### Why?
- One full orbit = 2π radians
- Earth completes orbit in 365 days

So:
```
angle = fraction_of_year * 2π
```

---

## Rogue Planet Position
```python
xA = 29.9 * math.cos(2 * math.pi * t / 1000)
yA = 100 * math.sin(2 * math.pi * t / 1000)
```

Same idea:
- Elliptical orbit
- Different size
- Slower period (1000 days)

---

## Why Calculate Distance This Way?

You want distance between two points:
```
Earth = (xE, yE)
Rogue = (xA, yA)
```

---

## Distance Formula

From geometry (Pythagorean theorem):
```
distance = √[(x1 - x2)² + (y1 - y2)²]
```

This comes from:
- horizontal difference = Δx
- vertical difference = Δy

The straight-line distance (hypotenuse) is:
```
√(Δx² + Δy²)
```

---

## Code
```python
return math.sqrt((xE - xA)**2 + (yE - yA)**2)
```

This is EXACTLY the Euclidean distance formula.

---

## Visual Intuition

Imagine:
```
Earth point   (xE, yE)
Rogue point   (xA, yA)
```

Draw a right triangle:
```
horizontal leg = xE - xA
vertical leg   = yE - yA
```

Distance = hypotenuse.

---

## Why Subtract Coordinates?

Because distance depends on relative position, not absolute position.

### Example
```
Earth at (10,5)
Rogue at (7,1)

Δx = 10 - 7 = 3
Δy = 5 - 1 = 4

distance = sqrt(3² + 4²) = 5
```

````python
# Asteroid.py

"""Illustrates how to find the minimum
amongst a number of function evaluations."""

import math

def dist(t):
    """Returns a float that is the distance between Earth and a rogue
    asteroid at time t (days)

    PreC: t is a nonnegative float."""

    xE = 94.5 * math.cos(2 * math.pi * t / 365)
    yE = 91.5 * math.sin(2 * math.pi * t / 365)
    xA = 29.9 * math.cos(2 * math.pi * t / 1000)
    yA = 100 * math.sin(2 * math.pi * t / 1000)

    return math.sqrt((xE - xA)**2 + (yE - yA)**2)


# Application Script
if __name__ == '__main__':
    """Find the minimum of dist(t) where t is an
    integer that satisfies L <= t <= R."""

    L = int(input('Enter initial time (integer): '))
    R = int(input('Enter final time (integer): '))

    # At any stage of the search, d_min is the smallest value of dist(t)
    # found thus far and t_min is the time associated with that minimum.
    d_min = dist(L)
    t_min = L

    for t in range(L + 1, R + 1):
        d_current = dist(t)
        if d_current < d_min:
            # A new minimum has been found.
            d_min = d_current
            t_min = t

    print(f"\n\nTime interval = [{L},{R}]")
    print(f"At t = {t_min}, the Earth-Asteroid distance is {(10**6)*d_min:5.0f} miles")
````

````python
# Zeno.py Zeno’s Paradox
"""Draws a sequence of disks along the x-axis.
The disks are tangent to each other and each disk has
a radius that is half as big as the one to its left."""

from SimpleGraphics import *

MakeWindow(8, bgcolor=BLACK)

rc = 4.0     # current radius
xc = -4.0    # current x-center
yc = 0.0     # y-center (constant)

for k in range(10):
    DrawDisk(xc, yc, rc, FillColor=YELLOW)

    # Determine the next xc and rc
    xc = xc + 1.5 * rc
    rc = rc / 2

ShowWindow()
````

````python
# Checkers.py
"""Looks at iteration in the context of drawing patterns.
In this case, the pattern is a checkerboard."""

from SimpleGraphics import *

def DrawRow(x0, y0, s, n, c1, c2):
    """Draws a horizontal row of s-by-s squares. The
    center of the leftmost square is (x0,y0) and the
    first two leftmost squares have colors c1 and c2
    respectively.

    PreC: x0, y0, and s are floats, n is a positive integer,
    c1 and c2 are rgb arrays."""

    # (xc,yc) is the center of the next square to draw.
    xc = x0
    yc = y0

    for k in range(n):
        # Draw the kth square
        if k % 2 == 0:
            # Even-index squares have color c1
            DrawRect(xc, yc, s, s, FillColor=c1, EdgeWidth=0)
        else:
            # Odd-index squares have color c2
            DrawRect(xc, yc, s, s, FillColor=c2, EdgeWidth=0)

        # Shift next square to the right
        xc = xc + s


if __name__ == '__main__':
    """Draws a checkerboard."""

    # The checkerboard is n-by-n and the squares are s-by-s
    n = 8
    s = 1

    # The lower left square has center at (x0,y0)
    x0 = -s * (n - 1) / 2.0
    y0 = -s * (n - 1) / 2.0

    # The checkerboard colors are c1 and c2
    c1 = RED
    c2 = WHITE

    MakeWindow((n + 2) * s / 2, bgcolor=BLACK, labels=True)

    # y is the vertical coordinate of the next row to draw
    y = y0

    for k in range(n):
        # Draw the kth row
        if k % 2 == 0:
            DrawRow(x0, y, s, n, c1, c2)
        else:
            DrawRow(x0, y, s, n, c2, c1)

        # Move up one square
        y = y + s

    ShowWindow()
````

````python
# ShowSqrt.py
"""Illustrates the effect of N on the accuracy
of a sqrt function where N is the number of
rectangle averagings."""

import math

def sqrt(x, N=5):
    """Returns a float that is an approximate square root
    of x. N is the number of rectangle averagings associated with
    the method.

    PreC: x is a nonnegative float and N is a positive int.
    """

    x = float(x)

    if x == 0:
        return 0

    L = x
    for k in range(N):
        L = (L + x / L) / 2

    return L


# Application Script
if __name__ == '__main__':
    print('\nLook at |math.sqrt(x) - sqrt(x,N)| for various N...')
    print('\n     x        math.sqrt(x)     N=5        N=6         N=7')
    print('----------------------------------------------------------------')

    for k in range(1, 11):
        x = 2**k
        yTrue = math.sqrt(x)

        y5 = sqrt(x)
        e5 = abs(y5 - yTrue)

        y6 = sqrt(x, 6)
        e6 = abs(y6 - yTrue)

        y7 = sqrt(x, 7)
        e7 = abs(y7 - yTrue)

        print(f"{x:10.0f}  {yTrue:12.6f}   {e5:10.2e}  {e6:10.2e}  {e7:10.2e}")
````
