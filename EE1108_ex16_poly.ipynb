Python OOP

Object-Oriented Programming (OOP) is a programming style where programs are organized using objects and classes. It models real-world entities and their behaviors.

	•	Class: A blueprint for creating objects

	•	Object: An instance of a class

	•	Attributes: Data stored in the object
  
	•	Methods: Functions that belong to the class

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def drive(self):
        print(self.brand, "is driving at", self.speed, "mph")

car1 = Car("Toyota", 60)
car1.drive()
```

Explanation

	•	Car is a class

	•	car1 is an object

	•	brand and speed are attributes
  
	•	drive() is a method

OOP helps make programs modular, reusable, and easier to maintain.

Example of Abstraction in Python

Python often uses abstract classes from the abc module.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

dog = Dog()
dog.sound()
```

Key idea
	•	Animal defines a common interface

	•	Each subclass implements its own version

Key Benefits of Abstraction
	•	Simplifies complex systems

	•	Improves code organization

	•	Enhances security by hiding internal details
  
	•	Allows flexible program design


## Polymorphism

Why Polymorphism: The purpose of polymorphism in Python is to allow different objects to use the same interface (method name) while performing different behaviors. This makes programs more flexible, reusable, and easier to maintain.

	•	The same method name speak() is used.

	•	Different objects respond differently to the same method call.

	•	The program treats all objects as Animal, but each subclass provides its own behavior.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

class Bird(Animal):
    def speak(self):
        print("Bird chirps")

animals = [Dog(), Cat(), Bird()]

for animal in animals:
    animal.speak()
```

Encapsulation in Python (Object-Oriented Programming)


Encapsulation is the concept of bundling data (variables) and methods (functions) together inside a class and restricting direct access to some of the object’s internal data.

The goal is to protect the internal state of an object and control how it is accessed or modified.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)

account.deposit(500)
print(account.get_balance())
```

Explanation

	•	__balance is a private variable.

	•	It cannot be accessed directly from outside the class.

	•	Users must interact through methods like deposit() or get_balance().

Why Encapsulation is Important

	1.	Data Protection
Prevents accidental modification of critical data.

	2.	Controlled Access
Changes to data happen through defined methods.

	3.	Code Flexibility
Internal implementation can change without affecting external code.
