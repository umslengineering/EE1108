## class

Rotation by angle θ:
*   $x' = x\cos\theta - y\sin\theta$
*   $y' = x\sin\theta + y\cos\theta$




```python
from math import sqrt, sin, cos, pi

class Point:
    """Represents a point in the plane."""

    def __init__(self, x, y):
        self.x = float(x)
        self.y = float(y)

    def __str__(self):
        return f"({self.x:6.3f},{self.y:6.3f})"

    def Dist(self, other):
        """Distance between two points."""
        return sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def Rotate(self, theta):
        """Rotate point counterclockwise about origin by theta degrees."""
        rad = pi * theta / 180.0
        c = cos(rad)
        s = sin(rad)

        return Point(self.x*c - self.y*s,
                     self.x*s + self.y*c)

    def Reflect(self):
        """Reflect point across the line y = x."""
        return Point(self.y, self.x)
```

```python
"""
PointClassInteractiveDemo.py
Python 3 ready interactive demonstration of Point class geometry transformations.

This version improves clarity and adds teaching-friendly annotations.
"""

# from ThePointClass import Point
from SimpleGraphics import *
from random import uniform as randu

# ----------------------------------------------------
# Visualization helpers
# ----------------------------------------------------

def ShowPoint(P, color=RED, label=None):
    """Draw a point and optionally display a label."""
    DrawDisk(P.x, P.y, 0.1, FillColor=color)
    if label is not None:
        DrawText(P.x + 0.15, P.y + 0.15, label)


def Midpoint(P1, P2):
    """Return midpoint of two points."""
    xm = (P1.x + P2.x) / 2.0
    ym = (P1.y + P2.y) / 2.0
    return Point(xm, ym)


def RandomPoint(Lx, Rx, Ly, Ry):
    """Return a random point inside a rectangle."""
    return Point(randu(Lx, Rx), randu(Ly, Ry))

# ----------------------------------------------------
# Main demonstration
# ----------------------------------------------------

if __name__ == '__main__':

    # Window size
    n = 3
    MakeWindow(n)

    # Draw coordinate grid
    #	The variable z moves from −n + 1 up to n − 1.
    # Each value of z represents a grid line position.
	  # This loop draws multiple lines spaced evenly across the grid.
    for z in range(-n + 1, n):
        #Drawing Vertical Lines,
        #Since x stays the same (z) and y changes,
        #this creates a Vertical line
        DrawLineSeg(z, -n, z, n) #x0, y0, x1, y1
        #Drawing Horizontal Lines
        #Here y stays the same (z) while x changes,
        #this creating a Horizontal line
        DrawLineSeg(-n, z, n, z)

    Title("Point Class Geometry Demo")

    # -----------------------------
    # Create two sample points
    # -----------------------------

    P = Point(1, 2)
    Q = Point(-2, -2)
    # print(P)
    # print(Q)

    ShowPoint(P, RED, "P")
    ShowPoint(Q, BLUE, "Q")

    # Distance annotation
    d = P.Dist(Q)
    DrawText(-n + 0.3, n - 0.4,
         "Distance(P,Q) = %.2f" % d)

    # Midpoint demonstration
    M = Midpoint(P, Q)
    ShowPoint(M, GREEN, "M")

    # Reflection across y=x
    S = P.Reflect()
    ShowPoint(S, PINK, "Reflect(P)")

    # Rotation demonstration (30 degrees)
    T = Q.Rotate(30)
    ShowPoint(T, CYAN, "Rotate(Q,30°)")

    # -----------------------------
    # Random point experiment
    # -----------------------------

    for k in range(10):
        R = RandomPoint(1, 2, -2, -1)
        ShowPoint(R, MAGENTA)

    ShowWindow()
```
