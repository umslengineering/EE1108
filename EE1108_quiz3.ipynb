#Find out the numbers multiples of 2, 3, 7 in the range from 1 to 100000


```python
N = 0

for k in range(1, 100_001):
    if k % 2 == 0 and k % 3 == 0 and k % 7 == 0:
        N = N + 1

print(N)
```
