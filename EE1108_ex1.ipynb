```python
def main():
    print("This code runs only when executed directly")

if __name__ == "__main__":
    main()
```

Why if __name__ == "__main__" Matters


*   Prevents code from running when imported
*   Allows files to act as both scripts and modules
*   Essential for large engineering and research projects
*   Widely used in scientific Python software

###Show Format

```python
# Old Fashion
""" Illustrates formatted print. You are
never tested on this stuff. It is just handy in
assignments when you want to display floats, ints,
and strings and want things to look nice.
"""

# The f-format

x = 1234.123456789
print('\nThe f-format (decimal notation)')
print('\n%m.nf means allocate m spaces overall with n to the right of the decimal.')
print('The number is right-justified within the space allocated.\n\n')
print('x = %16.3f' % x)
print('x = %16.6f' % x)
print('x = %16.9f' % x)

print('\n%-m.nf means allocate m spaces overall with n to the right of the decimal.')
print('The number is left-justified within the space allocated.\n\n')
print('x = %-16.3f' % x)
print('x = %-16.6f' % x)
print('x = %-16.9f' % x)

# The e-format

x = 1234.123456789
print('\nThe e-format (scientific notation)')
print('\n%m.ne means allocate m spaces overall with n to the right of the decimal.')
print('The number is right-justified within the space allocated.\n\n')
print('x = %16.3e' % x)
print('x = %16.6e' % x)
print('x = %16.9e' % x)

print('\n%-m.ne means allocate m spaces overall with n to the right of the decimal.')
print('The number is left-justified within the space allocated.\n\n')
print('x = %-16.3e' % x)
print('x = %-16.6e' % x)
print('x = %-16.9e' % x)

# The d-format (integers)

x = 123456789
print('\nThe d-format (for integers)')
print('\n%md means allocate m spaces overall.')
print('The number is right-justified within the space allocated.\n\n')
print('x = %10d' % x)
print('x = %13d' % x)
print('x = %16d' % x)

print('\n%-md means allocate m spaces overall.')
print('The number is left-justified within the space allocated.\n\n')
print('x = %-10d' % x)
print('x = %-13d' % x)
print('x = %-16d' % x)

# The s-format (strings)

x = 'quick'
print('\n%s is used for strings')
print('The %s brown fox is %s.' % (x, x))
print('%10s' % 'the')
print('%10s' % x)
print('%10s' % ('the ' + x))
```

```python
# New Fashion
"""Illustrates formatted printing using modern Python f-strings.
This is the recommended approach in current Python versions.
"""

# The f-format (decimal notation)

x = 1234.123456789
print("\nThe f-format (decimal notation)")
print("\n{m}.{n}f means allocate m spaces overall with n to the right of the decimal.")
print("The number is right-justified within the space allocated.\n")

print(f"x = {x:16.3f}")
print(f"x = {x:16.6f}")
print(f"x = {x:16.9f}")

print("\n-{m}.{n}f means left-justified within the space allocated.\n")

print(f"x = {x:<16.3f}")
print(f"x = {x:<16.6f}")
print(f"x = {x:<16.9f}")

# The e-format (scientific notation)

x = 1234.123456789
print("\nThe e-format (scientific notation)")
print("\n{m}.{n}e means allocate m spaces overall with n to the right of the decimal.")
print("The number is right-justified within the space allocated.\n")

print(f"x = {x:16.3e}")
print(f"x = {x:16.6e}")
print(f"x = {x:16.9e}")

print("\n-{m}.{n}e means left-justified within the space allocated.\n")

print(f"x = {x:<16.3e}")
print(f"x = {x:<16.6e}")
print(f"x = {x:<16.9e}")

# The d-format (integers)

x = 123456789
print("\nThe d-format (for integers)")
print("\n{m}d means allocate m spaces overall.")
print("The number is right-justified within the space allocated.\n")

print(f"x = {x:10d}")
print(f"x = {x:13d}")
print(f"x = {x:16d}")

print("\n-{m}d means left-justified within the space allocated.\n")

print(f"x = {x:<10d}")
print(f"x = {x:<13d}")
print(f"x = {x:<16d}")

# The s-format (strings)

x = "quick"
print("\n{s} is used for strings")
print(f"The {x} brown fox is {x}.")
print(f"{'the':>10s}")
print(f"{x:>10s}")
print(f"{'the ' + x:>10s}")
```

Surface Area Increase Calculations

```python
# SurfaceArea.py
"""Surface Area Increase

Computes (in three different ways) the increase
in surface area of a sphere when its radius is
increased a small amount.
"""

from math import pi

# -----------------------------
# Acquire the data
# -----------------------------
r = float(input("Enter the radius (kilometers): "))
inc = float(input("Enter the radius increase (millimeters): "))

# -----------------------------
# Unit conversions
# -----------------------------
# Radius increase in kilometers
dr = inc / 1e6

# -----------------------------
# Display header
# -----------------------------
print("\nCompute/Approximate the surface area increase three ways")
print(f"\nSphere radius   = {r:10.4f} kilometers")
print(f"Radius increase = {inc:10.4f} millimeters")
print("\nSurface area increase in square kilometers:")

# -----------------------------
# Compute area increase (three methods)
# -----------------------------
del_A1 = 4 * pi * ((r + dr)**2 - r**2)
print(f"\n   Method 1: {del_A1:15.12f}")

del_A2 = 4 * pi * (2*r + dr) * dr
print(f"   Method 2: {del_A2:15.12f}")

del_A3 = 8 * pi * r * dr
print(f"   Method 3: {del_A3:15.12f}")
```

### Windchill Example

```python
# Windchill.py
"""Windchill is a function of temperature and wind speed.

This program solicits temperature and wind speed via input
and then displays the associated windchill temperature.
"""

# -----------------------------
# Acquire input (convert to float!)
# -----------------------------
temp = float(input("Enter a temperature ≤ 50°F: "))
wind = float(input("Enter a wind speed ≥ 3 mph: "))

# -----------------------------
# Model parameters
# -----------------------------
A = 35.74
B = 0.6215
C = -35.75
D = 0.4275
e = 0.16

# -----------------------------
# Compute windchill
# -----------------------------
WC = (A + B * temp) + (C + D * temp) * wind**e

# -----------------------------
# Display results
# -----------------------------
print(f"\nTemperature: {temp:.1f} °F")
print(f"Wind Speed : {wind:.1f} mph")
print(f"Windchill  : {WC:4.0f} °F")
```
