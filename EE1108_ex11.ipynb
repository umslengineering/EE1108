```python
def inS(a, b, x, y):
    """Returns True if (x,y) is inside the square with vertices
    (a,b), (a+1,b), (a,b+1), and (a+1,b+1).
    Otherwise, returns False."""

    xOK = (a <= x <= a + 1)
    yOK = (b <= y <= b + 1)
    z = (xOK and yOK)
    return z


a1 =2
a2=6.2
b1=3
b2=3.2
x=2.5
y=3.5
z_union = inS(a1, b1, x, y) or inS(a2, b2, x, y)
z_intersection = inS(a1, b1, x, y) and inS(a2, b2, x, y)
print(z_union, z_intersection)
```

```python
# ShowAssert.py
"""Demonstrates how assert statements can be used to check multiple preconditions."""

import string


def TimeToMinutes(T):
    """
    Returns the total number of minutes that have elapsed from midnight
    to the time specified by T.

    Precondition:
        T is a string in the format 'hh:mm'
        where hh is between 00 and 23
        and mm is between 00 and 59.

    Valid examples: '00:00', '12:34', '23:59'
    Invalid examples: '1:23', '24:01', '01:78', 'xy:zw'
    """

    # Validate preconditions
    assert isinstance(T, str), "T must be a string."
    assert len(T) == 5, "Time string must have length 5."
    assert T[2] == ":", "The third character must be a colon."
    assert T[0] in string.digits and T[1] in string.digits, \
        "The hour portion must be numeric."
    assert T[3] in string.digits and T[4] in string.digits, \
        "The minute portion must be numeric."

    hours = int(T[0:2])
    minutes = int(T[3:5])

    assert 0 <= hours <= 23, "Hour must satisfy 0 <= hh <= 23."
    assert 0 <= minutes <= 59, "Minute must satisfy 0 <= mm <= 59."

    total_minutes = 60 * hours + minutes

    # Postcondition check
    assert isinstance(total_minutes, int), "Return value must be an integer."

    return total_minutes


def MinutesToTime(M):
    """
    Returns a time string in the format 'hh:mm' representing
    M minutes past midnight.

    Precondition:
        M is an integer satisfying 0 <= M < 24*60.
    """

    # Validate preconditions
    assert isinstance(M, int), "M must be an integer."
    assert 0 <= M < 24 * 60, "M must satisfy 0 <= M < 1440."

    # Compute hours and minutes
    hours = M // 60
    minutes = M % 60

    # Format with leading zeros
    T = f"{hours:02d}:{minutes:02d}"

    # Postcondition validation
    assert isinstance(T, str), "Return value must be a string."
    assert len(T) == 5, "Time string must have length 5."
    assert T[2] == ":", "The third character must be a colon."
    assert T[0] in string.digits and T[1] in string.digits, \
        "Hour portion must be numeric."
    assert T[3] in string.digits and T[4] in string.digits, \
        "Minute portion must be numeric."

    return T


# Application script
if __name__ == "__main__":
    """
    Prompts the user for a time string, converts it to minutes,
    and then converts the minutes back to a time string.
    """

    T = input("Enter a time string in the form hh:mm: ")
    print("Input time:", T)

    M = TimeToMinutes(T)
    print("Total minutes:", M)

    T = MinutesToTime(M)
    print("Converted back to time:", T)
```

```python
# ShowBreak.py
"""
Illustrates the use of the break command.
Prompts the user to enter a nonnegative integer,
which is required by the factorial function.
"""

import math

while True:
    try:
        N = input("Enter a nonnegative integer: ")
        N = int(N)

        if N >= 0:
            # Valid input — exit loop
            break
        else:
            print("N must be nonnegative.")

    except ValueError:
        print("Invalid input. Please enter an integer.")

# Compute factorial
M = math.factorial(N)

print(f"\n\n{N}! = {M}")
```

```python
"""
ShowTry2.py
Illustrates the use of try-except.
Encourages the user to input a nonnegative integer because
that is what the factorial function requires
"""

import math

while True:
    N = input("Enter a nonnegative integer: ")

    try:
        # Convert the input string to an int.
        N = int(N)

        if N < 0:
            print("N must be nonnegative.")
            continue

        # Valid input. Terminate the loop.
        break

    except ValueError:
        # Invalid input. Inform the user and the iteration continues.
        print("N must have type int")

M = math.factorial(N)

print(f"\n\n{N}! = {M}")
```

```python
"""
SqrtAssert.py
Shows how the assert statement can be used to check
that the preconditions of a function are satisfied
and that the returned value satisfies the specification.
"""

def sqrt(x):
    """
    Returns an approximate square root L of x such that
    |L*L - x| <= 0.001

    Precondition:
        x is a positive number.
    """

    assert x > 0, "The sqrt function requires a positive argument."

    L = float(x)

    # Newton's method iteration (4 iterations as in original code)
    L = (L + x / L) / 2
    L = (L + x / L) / 2
    L = (L + x / L) / 2
    L = (L + x / L) / 2

    assert abs(L * L - x) <= 0.001, "Inaccurate Square Root"

    return L


if __name__ == "__main__":
    x = float(input("Enter a positive number: "))
    z = sqrt(x)

    print(f"sqrt({x:5.3f}) = {z:10.6f}")
```
