```python
# GetData.py
""" This module provides functions that can be
used to extract data from text files.
"""

def fileToStringList(FName):
    """Returns a list of strings L with the property that
    L[k] is the kth line in the file with name FName.
    \r: It moves the cursor to the beginning of the current line without moving to the next line.
	  It is often used together with \n (newline).
    """
    L = []
    with open(FName, "r", encoding="latin-1") as F:
        for s in F:
            s = s.rstrip("\n\r")
            L.append(s)
    return L


def stringToWordList(s):
    """Returns t.split() where t is obtained from s.lower() by replacing
    every non-letter in this string with a blank. Basically, it returns
    a list of the words in s.

    PreC: s is a string.
    """
    t = ""
    for c in s.lower():
        if c.isalpha():
            t += c
        else:
            t += " "
    return t.split()
```

```python
# TheCountyPopClass.py
# __init__ is a special method called a constructor.
# It runs automatically whenever you create a new instance of the class.
# self refers to the current instance of the class.
# Parameters:
# name → name of the county (string)
# state → state where the county is located (string)
# pop2010 … pop2014 → population of the county from 2010 to 2014 (integers)
# -> None → indicates that this method does not return anything (Python type hint).

class CountyPop:
    def __init__(
        self,
        name: str,
        state: str,
        pop2010: int,
        pop2011: int,
        pop2012: int,
        pop2013: int,
        pop2014: int,
    ) -> None:
        #Assigning attributes
        self.name = name
        self.state = state
        self.pop2010 = int(pop2010)
        self.pop2011 = int(pop2011)
        self.pop2012 = int(pop2012)
        self.pop2013 = int(pop2013)
        self.pop2014 = int(pop2014)

    def __str__(self) -> str:
        full_name = f"{self.name}, {self.state}"
        return f"{full_name:>40}"

# This means:

# :>40 → right-align the text
# in a field of width 40 characters
```
